```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_excel('/content/overalldata.xlsx')

stats.probplot(df['Flesch Reading Ease'], dist="norm", plot=plt)
plt.title("Q-Q Plot for Flesch Reading Ease")
plt.show()
```
