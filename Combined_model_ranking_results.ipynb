# # Top-performing models in comparisson to original text results.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

file_path = '/content/combined_model_ranking_all.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')

def create_radar_chart(df, model_names, rank_columns):

    categories = rank_columns
    N = len(categories)

    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    angles += angles[:1]

    plt.figure(figsize=(8, 8))

    for model in model_names:
        values = df[df['model'] == model][rank_columns].mean().values.flatten().tolist()
        values += values[:1]

        plt.polar(angles, values, label=model)
        plt.fill(angles, values, alpha=0.25)

    plt.xticks(angles[:-1], categories, color='grey', size=8)
    plt.yticks([5, 10, 15, 20], ["5", "10", "15", "20"], color="grey", size=7)
    plt.ylim(0, 20)

    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.title('Rank Comparison Across Metrics')
    plt.show()

rank_columns = [
    'Automated Readability Index Rank',
    'Coleman-Liau Index Rank',
    'Dale-Chall Readability Score Rank',
    'Flesch Reading Ease Rank',
    'Flesch-Kincaid Grade Level Rank',
    'Gunning Fog Rank',
    'Lexicon Count Rank',
    'Linsear Write Formula Rank',
    'Reading Time (seconds) Rank',
    'SMOG Index Rank',
    'Sentence Count Rank',
    'Syllable Count Rank'
]

comparison = data[['model', 'Final Rank'] + rank_columns].dropna()
top_model_names = comparison['model'].head(3).tolist()

create_radar_chart(data, top_model_names, rank_columns)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

file_path = '/content/combined_model_ranking_all.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')

rank_columns = [
    'Automated Readability Index Rank',
    'Coleman-Liau Index Rank',
    'Dale-Chall Readability Score Rank',
    'Flesch Reading Ease Rank',
    'Flesch-Kincaid Grade Level Rank',
    'Gunning Fog Rank',
    'Lexicon Count Rank',
    'Linsear Write Formula Rank',
    'Reading Time (seconds) Rank',
    'SMOG Index Rank',
    'Sentence Count Rank',
    'Syllable Count Rank'
]

worst_three_models = data[['model'] + rank_columns + ['Final Rank']].sort_values(by='Final Rank', ascending=False).head(3)

def create_radar_chart(df, model_names, rank_columns):
    categories = rank_columns
    N = len(categories)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    angles += angles[:1]

    plt.figure(figsize=(8, 8))

    for model in model_names:
        values = df[df['model'] == model][rank_columns].mean().values.flatten().tolist()
        values += values[:1]

        plt.polar(angles, values, label=model)
        plt.fill(angles, values, alpha=0.25)

    plt.xticks(angles[:-1], categories, color='grey', size=8)
    plt.yticks([5, 10, 15, 20], ["5", "10", "15", "20"], color="grey", size=7)
    plt.ylim(0, 20)

    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.title('Rank Comparison Across Metrics')
    plt.show()

worst_model_names = worst_three_models['model'].tolist()
create_radar_chart(data, worst_model_names, rank_columns)
```
