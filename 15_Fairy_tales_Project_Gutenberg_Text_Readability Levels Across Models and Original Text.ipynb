```python
import pandas as pd

file_path = '/content/overalldata.xlsx'


df = pd.read_excel(file_path, sheet_name='Sheet1')

print(df.head())

mean_values = df.mean(numeric_only=True)
print("Mean values for each column:")
print(mean_values)

highest_flesch_ease_model = df[df['Flesch Reading Ease'] == df['Flesch Reading Ease'].max()]
print("\nModel with the highest Flesch Reading Ease score:")
print(highest_flesch_ease_model)

lowest_flesch_ease_model = df[df['Flesch Reading Ease'] == df['Flesch Reading Ease'].min()]
print("\nModel with the lowest Flesch Reading Ease score:")
print(lowest_flesch_ease_model)

comparison_all_models = df.set_index('model')
print("\nMetrics for each model:")
print(comparison_all_models)

comparison_all_models_display = comparison_all_models[[
    'Automated Readability Index', 'Coleman-Liau Index', 'Dale-Chall Readability Score',
    'Flesch Reading Ease', 'Flesch-Kincaid Grade Level', 'Gunning Fog', 'Lexicon Count',
    'Linsear Write Formula', 'Reading Time (seconds)', 'SMOG Index', 'Sentence Count',
    'Syllable Count'
]]
comparison_all_models_display
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/overalldata.xlsx'
data = pd.read_excel(file_path, sheet_name='Sheet1')

readability_order = {
    '2nd and 3rd grade': 1,
    '4th and 5th grade': 2,
    '5th and 6th grade': 3,
    '6th and 7th grade': 4,
    '7th and 8th grade': 5,
    '8th and 9th grade': 6,
    '9th and 10th grade': 7,
    '10th and 11th grade': 8,
    '11th and 12th grade': 9,
}

def get_readability_levels(model_name):
    model_text_standard = data[data['model'] == model_name][['Text Standard']]
    model_text_standard = model_text_standard.dropna().reset_index(drop=True)
    return model_text_standard

# readability levels extraction for the original text
original_text_levels = get_readability_levels('original_text')['Text Standard'].str.split(', ', expand=True).stack().reset_index(drop=True)
original_text_counts = original_text_levels.value_counts().reindex(readability_order.keys()).fillna(0)

# each model readability levels extraction
model_readability_counts = {}
for model in data['model'].unique():
    if model != 'original_text':
        model_levels = get_readability_levels(model)['Text Standard'].str.split(', ', expand=True).stack().reset_index(drop=True)
        model_counts = model_levels.value_counts().reindex(readability_order.keys()).fillna(0)
        model_readability_counts[model] = model_counts

readability_df = pd.DataFrame(model_readability_counts)
readability_df['Original Text'] = original_text_counts
readability_df = readability_df.T


plt.figure(figsize=(15, 10))
sns.barplot(data=readability_df)
plt.title('Readability Levels Across Models and Original Text')
plt.xlabel('Readability Level')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.legend(title='Models')
plt.tight_layout()
plt.show()
```
